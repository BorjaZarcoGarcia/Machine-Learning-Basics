# 🌐 Guía de Visualizaciones Interactivas con Plotly

## 🎯 Introducción

**Plotly** es una librería de visualización moderna que crea **gráficos interactivos** listos para la web. A diferencia de Matplotlib y Seaborn (que generan imágenes estáticas), Plotly permite:

- ✨ **Interactividad**: Zoom, pan, hover sobre puntos para ver datos
- 🌐 **Exportación web**: Gráficos embebibles en páginas HTML
- 🎨 **Diseño profesional**: Estética moderna y limpia por defecto
- 📱 **Responsividad**: Se adaptan automáticamente al tamaño de pantalla

**Plotly Express** (`px`) es la API de alto nivel que hace que crear visualizaciones sea tan simple como con Seaborn, pero con resultados interactivos.

# 📥 1. Carga de Datos

Cargaremos cuatro datasets diferentes para explorar las capacidades de Plotly.

```python
# Importar librerías
import pandas as pd
import plotly.express as px  # API de alto nivel de Plotly

print("📥 Cargando datasets...")

# Dataset 1: Pingüinos de la Antártida
penguins = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/penguins.csv')
print(f"✅ Penguins: {penguins.shape[0]} filas, {penguins.shape[1]} columnas")

# Dataset 2: Transporte público de Chicago (CTA)
cta = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/CTA_-_Ridership_-_Daily_Boarding_Totals.csv')
cta['service_date'] = pd.to_datetime(cta['service_date'])
print(f"✅ CTA Chicago: {cta.shape[0]} filas, {cta.shape[1]} columnas")

# Dataset 3: Titanic
titanic = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/titanic.csv')
print(f"✅ Titanic: {titanic.shape[0]} filas, {titanic.shape[1]} columnas")

# Dataset 4: 50 Startups
startups = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/50_Startups.csv')
print(f"✅ 50 Startups: {startups.shape[0]} filas, {startups.shape[1]} columnas")

print("\n🎉 Todos los datasets cargados correctamente")
```

# 📊 2. Gráficos Básicos con Plotly Express

Plotly Express (`px`) permite crear gráficos interactivos con muy poco código.

### Scatter Plots (Gráficos de Dispersión)

Muestran la relación entre dos variables numéricas. Con Plotly, puedes hacer zoom, pan y hover sobre los puntos.

```python
# Scatter plot básico con colores por especie
fig = px.scatter(penguins,
                 x='bill_length_mm',
                 y='bill_depth_mm',
                 color='species',
                 # También se pueden colorear los puntos según un atributo numérico, p.ej. 'flipper_length_mm'
                 # Opcional:
                 # hover_data=['flipper_length_mm', 'body_mass_g'],
                 # hover_name='island',
                 title='Relación entre longitud y profundidad del pico por especie')
fig.show()

print("💡 Pasa el cursor sobre los puntos para ver detalles. Usa los controles de zoom y pan.")
```

### Histogramas

Visualizan la distribución de datos categóricos o numéricos.

```python
# Histograma básico
fig = px.histogram(penguins,
                   x='island',
                   labels={'island':'Nombre de la Isla'},
                   # OPCIONAL:
                   # color='species',
                   # barmode='group', # las opciones son 'stack' y 'group'
                   title='Distribución de pingüinos por isla')
fig.show()

print("💡 Plotly cuenta automáticamente las categorías, igual que Seaborn")
```

### Gráficos de Líneas (Series Temporales)

Perfectos para visualizar evolución de datos en el tiempo con interactividad.

```python
# Filtrar datos desde 2018
start_date = pd.to_datetime('2018-01-01')
cta_recent = cta[cta['service_date'] >= start_date].copy()
cta_recent = cta_recent.sort_values('service_date')

# Gráfico de líneas
fig = px.line(cta_recent,
              x='service_date',
              y='total_rides',
              title='Uso diario del transporte público en Chicago (desde 2018)')
fig.show()

print("💡 ¡Puedes hacer zoom en la caída de 2020 (COVID-19) para ver más detalle!")
```

### Box Plots (Diagramas de Caja)

Muestran la distribución estadística de datos con mediana, cuartiles y outliers.

```python
# Box plot básico
fig = px.box(titanic,
             x='Pclass',
             y='Age',
             title='Distribución de edades por clase en el Titanic')
fig.show()

print("💡 Haz clic en las cajas para ver estadísticas detalladas")
```

# 🎨 3. Personalización Avanzada

Plotly permite agregar múltiples dimensiones de información en un solo gráfico usando color, tamaño, hover data y más.

### Box Plot con Múltiples Variables

Combinar color, hover data y labels personalizados.

```python
# Box plot con color por género y hover data adicional
fig = px.box(titanic,
             x='Pclass',
             y='Age',
             color='Sex',
             labels={'Pclass': 'Clase Social', 'Age': 'Edad'},
             hover_data=['SibSp', 'Parch'],
             title='Distribución de edades por clase y género en el Titanic')
fig.show()

print("💡 El parámetro 'color' divide cada caja por categoría")
print("🔹 'hover_data' agrega variables extra al pasar el cursor")
print("🔹 'labels' personaliza los nombres de los ejes")
```

### Scatter Plot Multidimensional

Visualizar hasta 5 dimensiones en un solo gráfico: X, Y, color, tamaño y hover.

```python
# Scatter plot con 5 dimensiones de información
fig = px.scatter(startups,
                 x='R&D Spend',
                 y='Profit',
                 color='State',              # Dimensión 3: Color por estado
                 size='Marketing Spend',     # Dimensión 4: Tamaño por gasto en marketing
                 hover_name='State',
                 hover_data=['Administration'],  # Dimensión 5: Info adicional en hover
                 size_max=20,
                 title='Gasto en I+D vs Ganancias (tamaño = Marketing, color = Estado)')

fig.update_layout(width=800)
fig.show()

print("💡 Este gráfico muestra 5 dimensiones simultáneamente:")
print("  🔹 X: Gasto en I+D")
print("  🔹 Y: Ganancias")
print("  🔹 Color: Estado")
print("  🔹 Tamaño: Gasto en Marketing")
print("  🔹 Hover: Gastos de Administración")
```

**Cuándo usar Plotly:**
- ✅ Dashboards interactivos
- ✅ Presentaciones web
