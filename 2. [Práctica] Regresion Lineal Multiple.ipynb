# 📊 Regresión Lineal Múltiple: Predicción de Rentabilidad de Startups

## 🎯 Introducción

La **Regresión Lineal Múltiple** es una extensión natural de la regresión lineal simple que permite **predecir una variable continua** usando **múltiples variables predictoras** simultáneamente.

**En este notebook** aprenderás a:
- Trabajar con múltiples variables predictoras
- Dividir datos en conjuntos de entrenamiento y prueba
- Entrenar un modelo de regresión lineal múltiple
- Evaluar el rendimiento con RMSE
- Comparar modelos con diferentes conjuntos de features
- Detectar overfitting comparando errores de train vs test

### 💼 Caso de Uso: Predicción de Rentabilidad de Startups

Usaremos el dataset **50 Startups** que contiene información de 50 empresas emergentes con:
- **R&D Spend**: Gasto en investigación y desarrollo ($)
- **Administration**: Gastos administrativos ($)
- **Marketing Spend**: Gasto en marketing ($)
- **State**: Estado donde opera (categórica) - *la omitiremos en este análisis*
- **Profit** (target): Rentabilidad de la empresa ($)

**Objetivo**: Construir un modelo que **prediga la rentabilidad** de una startup basándose en sus gastos en diferentes áreas.

## 📐 Fundamentos Matemáticos

La regresión lineal múltiple extiende la ecuación de la recta a múltiples dimensiones:

$$y = b_0 + b_1x_1 + b_2x_2 + b_3x_3 + ... + b_nx_n$$

Donde:
- **y**: Variable a predecir (Profit)
- **x₁, x₂, x₃**: Variables predictoras (R&D, Admin, Marketing)
- **b₀**: Intercepto (rentabilidad base)
- **b₁, b₂, b₃**: Coeficientes - contribución de cada variable al profit

El algoritmo encuentra los valores óptimos de todos los coeficientes minimizando el error cuadrático.

## 💡 Regresión Simple vs Múltiple

| Aspecto | Regresión Simple | Regresión Múltiple |
|---------|------------------|-------------------|
| **Variables predictoras** | 1 (univariada) | Múltiples (multivariada) |
| **Ecuación** | y = mx + n | y = b₀ + b₁x₁ + b₂x₂ + ... |
| **Interpretación** | Muy fácil | Requiere análisis de coeficientes |
| **Poder predictivo** | Limitado | Mayor (captura relaciones complejas) |
| **Riesgo de overfitting** | Bajo | Mayor (más parámetros) |

**Ventajas de Regresión Múltiple**:
- ✅ Mayor poder predictivo
- ✅ Captura interacciones entre variables
- ✅ Más realista para problemas reales

**Desafíos**:
- ⚠️ Multicolinealidad (variables correlacionadas)
- ⚠️ Mayor riesgo de overfitting
- ⚠️ Interpretación más compleja

# 📥 1. Carga y Exploración de Datos

Comenzamos cargando el dataset de 50 Startups y realizando una exploración inicial.

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("📥 Cargando dataset de 50 Startups...")

# Cargar dataset desde GitHub
url = "https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/50_Startups.csv"
df = pd.read_csv(url)

print(f"✅ Dataset cargado correctamente")
print(f"\n📊 Dimensiones del dataset: {df.shape[0]} filas, {df.shape[1]} columnas")

# Explorar primeras filas
print("\n🔹 Primeras filas del dataset:")
display(df.head())

print("\n💡 Variables del dataset:")
print("  🔹 R&D Spend: Gasto en I+D ($)")
print("  🔹 Administration: Gastos administrativos ($)")
print("  🔹 Marketing Spend: Gasto en marketing ($)")
print("  🔹 State: Estado de operación (categórica)")
print("  🔹 Profit: Rentabilidad (target) ($)")
```

```python
# Eliminar variable categórica 'State'
# En este ejercicio nos enfocamos solo en variables numéricas
df_num = df.drop('State', axis=1)

print("🔹 Dataset después de eliminar 'State':")
display(df_num.head())

print(f"\n✅ Variables restantes: {list(df_num.columns)}")
print(f"📊 Todas las variables son numéricas: {df_num.shape[1]} columnas")
```

# 🔀 2. Preparación de Datos

Separamos las variables predictoras del objetivo y dividimos los datos en conjuntos de entrenamiento y prueba.

## ¿Por qué dividir en Train/Test?

- **Train (80%)**: Datos que el modelo "ve" durante el entrenamiento
- **Test (20%)**: Datos que el modelo NO ve - simula datos futuros
- **Objetivo**: Evaluar cómo generaliza el modelo a datos nuevos

```python
# Separar variables predictoras (X) y objetivo (y)
X = df_num.drop('Profit', axis=1)  # Variables de entrada: R&D, Admin, Marketing
y = df_num['Profit']                # Variable objetivo: Profit

print("🔹 Variables predictoras (X):")
display(X.head())

print(f"\n🔹 Variable objetivo (y): Profit")
print(y.head().values)

# División train/test
x_train, x_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% para test
    random_state=42     # Para reproducibilidad
)

print(f"\n📊 División completada:")
print(f"  🔹 Train: {x_train.shape[0]} muestras ({x_train.shape[0]/len(X)*100:.0f}%)")
print(f"  🔹 Test: {x_test.shape[0]} muestras ({x_test.shape[0]/len(X)*100:.0f}%)")
print(f"  🔹 Features: {x_train.shape[1]} variables predictoras")

print(f"\n💡 random_state=42 garantiza que la división sea siempre igual")
```

# 🤖 3. Entrenamiento del Modelo

Entrenamos un modelo de **Regresión Lineal Múltiple** usando scikit-learn.

## ¿Qué hace el modelo?

El modelo busca la ecuación óptima:

$$Profit = b_0 + b_1 \cdot RD + b_2 \cdot Admin + b_3 \cdot Marketing$$

Donde minimiza el error cuadrático entre las predicciones y los valores reales.

```python
# Crear modelo de regresión lineal múltiple
lin_reg = LinearRegression()

# Entrenar el modelo con los datos de train
lin_reg.fit(x_train, y_train)

print("✅ Modelo entrenado exitosamente\n")

# Examinar los coeficientes aprendidos
print("📐 Ecuación del modelo:")
print(f"\n   Profit = {lin_reg.intercept_:.2f}")

for i, col in enumerate(x_train.columns):
    signo = "+" if lin_reg.coef_[i] >= 0 else ""
    print(f"            {signo} {lin_reg.coef_[i]:.2f} × {col}")

print(f"\n💡 Interpretación de coeficientes:")
print(f"  🔹 Intercepto (b₀): {lin_reg.intercept_:.2f}")
print(f"     → Profit base cuando todos los gastos son 0")

for i, col in enumerate(x_train.columns):
    print(f"  🔹 {col}: {lin_reg.coef_[i]:.2f}")
    if lin_reg.coef_[i] > 0:
        print(f"     → Por cada $1 extra en {col}, el profit aumenta ${lin_reg.coef_[i]:.4f}")
    else:
        print(f"     → Por cada $1 extra en {col}, el profit disminuye ${abs(lin_reg.coef_[i]):.4f}")

# Hacer predicciones sobre el conjunto de test
y_predictions = lin_reg.predict(x_test)

print(f"\n🔮 Predicciones realizadas sobre {len(y_predictions)} muestras de test")
```

# 📊 4. Evaluación del Rendimiento

Evaluamos qué tan bien predice el modelo usando varias métricas de error.

## Métricas de Evaluación

- **MAE** (Mean Absolute Error): Error promedio en valor absoluto
- **RMSE** (Root Mean Squared Error): Penaliza más los errores grandes
- **R²** (Coeficiente de Determinación): % de varianza explicada (0-1)

```python
# Calcular métricas de evaluación
mae = mean_absolute_error(y_test, y_predictions)
mse = mean_squared_error(y_test, y_predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_predictions)

print("📊 Rendimiento del Modelo (sobre datos de Test):\n")
print(f"  🔹 MAE:  ${mae:,.2f}")
print(f"  🔹 RMSE: ${rmse:,.2f}")
print(f"  🔹 R²:   {r2:.4f} ({r2*100:.2f}%)\n")

print("💡 Interpretación:")
print(f"  • En promedio, el modelo se equivoca por ${mae:,.2f}")
print(f"  • El modelo explica el {r2*100:.1f}% de la varianza en Profit")
print(f"  • RMSE de ${rmse:,.2f} indica el error típico de las predicciones\n")

# Mostrar algunas predicciones vs valores reales
print("🔍 Comparación de algunas predicciones:\n")
comparison_df = pd.DataFrame({
    'Valor Real': y_test.values[:5],
    'Predicción': y_predictions[:5],
    'Error': y_test.values[:5] - y_predictions[:5],
    'Error %': ((y_test.values[:5] - y_predictions[:5]) / y_test.values[:5] * 100)
})
display(comparison_df)
```

# 🔍 5. Selección de Features

Experimentamos entrenando modelos con **subconjuntos de variables** para comparar rendimiento.

## ¿Más variables = Mejor modelo?

No necesariamente. A veces:
- Algunas variables tienen poco poder predictivo
- Menos variables = modelo más simple y generalizable
- Evitamos multicolinealidad (variables correlacionadas)

**Experimento**: Entrenar un modelo usando solo **2 de las 3 variables** disponibles.

```python
# Experimento: Modelo con solo 2 features (Administration y Marketing Spend)
# Omitimos R&D Spend para ver cómo afecta al rendimiento

print("🧪 Experimento: Entrenar modelo con solo 2 features\n")
print("   Omitimos: R&D Spend")
print("   Usamos: Administration y Marketing Spend\n")

# Seleccionar solo 2 features
features_reducidas = ["Administration", "Marketing Spend"]
x_train_new = x_train[features_reducidas]
x_test_new = x_test[features_reducidas]

print(f"📊 Dimensiones de datos reducidos:")
print(f"  🔹 Train: {x_train_new.shape}")
print(f"  🔹 Test: {x_test_new.shape}")

# Entrenar nuevo modelo
lin_reg_new = LinearRegression()
lin_reg_new.fit(x_train_new, y_train)

print(f"\n✅ Modelo reducido entrenado")

# Hacer predicciones
y_predictions_new = lin_reg_new.predict(x_test_new)

# Evaluar nuevo modelo
mae_new = mean_absolute_error(y_test, y_predictions_new)
mse_new = mean_squared_error(y_test, y_predictions_new)
rmse_new = np.sqrt(mse_new)
r2_new = r2_score(y_test, y_predictions_new)

print(f"\n📊 Rendimiento del Modelo Reducido (2 features):\n")
print(f"  🔹 MAE:  ${mae_new:,.2f}")
print(f"  🔹 RMSE: ${rmse_new:,.2f}")
print(f"  🔹 R²:   {r2_new:.4f} ({r2_new*100:.2f}%)\n")

# Comparar con el modelo completo
print("🔎 Comparación: Modelo Completo vs Reducido\n")
comparison = pd.DataFrame({
    'Métrica': ['MAE', 'RMSE', 'R²'],
    'Modelo Completo (3 features)': [f"${mae:,.2f}", f"${rmse:,.2f}", f"{r2:.4f}"],
    'Modelo Reducido (2 features)': [f"${mae_new:,.2f}", f"${rmse_new:,.2f}", f"{r2_new:.4f}"]
})
display(comparison)

print(f"\n💡 Conclusión:")
if r2_new < r2:
    perdida = (r2 - r2_new) * 100
    print(f"  • El modelo reducido pierde {perdida:.1f} puntos porcentuales de R²")
    print(f"  • Esto sugiere que R&D Spend es importante para predecir Profit")
else:
    print(f"  • El modelo reducido mantiene buen rendimiento")
    print(f"  • Quizás R&D Spend no aporta mucha información adicional")
```

# ⚠️ 6. Detección de Overfitting

El **overfitting** (sobreajuste) ocurre cuando el modelo aprende demasiado bien los datos de entrenamiento, pero no generaliza bien a datos nuevos.

## ¿Cómo detectar overfitting?

Comparamos el error del modelo en:
- **Train**: Datos que el modelo vio durante el entrenamiento
- **Test**: Datos que el modelo NO vio

**Señal de overfitting**: Error mucho menor en train que en test

## Criterio de detección

Si el error en train es significativamente menor (por ejemplo, <95% del error de test), podría haber overfitting.
