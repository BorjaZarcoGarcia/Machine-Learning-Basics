# ⚡ Guía de Polars: La Alternativa Ultrarrápida a Pandas

## 🎯 Introducción

**Polars** es una librería moderna de manipulación de datos escrita en Rust que ofrece un rendimiento excepcional. Es una alternativa a Pandas diseñada específicamente para velocidad y eficiencia en el procesamiento de grandes volúmenes de datos.

### 🔥 ¿Por qué Polars?

- **⚡ Velocidad excepcional**: Hasta 10-100x más rápido que Pandas en operaciones complejas
- **🧠 Evaluación perezosa**: Optimiza automáticamente las consultas antes de ejecutarlas
- **💪 Manejo eficiente de memoria**: Diseñado para trabajar con grandes datasets
- **🎨 Sintaxis expresiva**: API clara y encadenable, más legible que Pandas
- **🔧 Paralelización automática**: Usa todos los núcleos de CPU disponibles

## 🔧 Operaciones principales que veremos en este notebook:

### **📥 1. Carga de Datos y Exploración Básica**
- `pl.read_csv()` - Cargar datos desde archivos CSV o URLs
- Comparación de sintaxis con Pandas
- Exploración de dimensiones y estructura

### **🎯 2. Selección y Filtrado de Datos**
- `df.filter()` - Filtrado de filas con condiciones
- `df.select()` - Selección de columnas
- `pl.col()` - Referencia a columnas de forma explícita
- Comparación de legibilidad con Pandas

### **⚡ 3. Comparativa de Rendimiento**
- Operaciones de agrupación y agregación
- `df.group_by()` y `agg()` - Agrupamiento eficiente
- Medición de tiempos de ejecución
- Análisis de velocidad relativa

### **🚀 4. Prueba de Estrés con Grandes Volúmenes**
- Manejo de millones de filas
- `pl.concat()` - Concatenación de DataFrames
- Comparación de rendimiento a gran escala

### **🧠 5. Evaluación Perezosa (Lazy Evaluation)**
- `pl.scan_csv()` - Lectura perezosa de datos
- `LazyFrame` vs `DataFrame`
- `.collect()` - Ejecución optimizada de consultas
- Optimización automática de planes de ejecución

## ⚙️ Configuración Inicial

Importamos las librerías necesarias para trabajar con Polars y compararlo con Pandas.

```python
import time
import pandas as pd
import polars as pl

print("✅ Librerías importadas correctamente")
print(f"📦 Versión de Polars: {pl.__version__}")
print(f"📦 Versión de Pandas: {pd.__version__}")
```

# 📥 1. Carga de Datos y Exploración Básica

Cargaremos el mismo dataset en Polars y Pandas para comparar sintaxis y funcionalidad básica.

```python
# URL del dataset de 100,000 clientes
csv_url = "https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/customers-100000.csv"

# Cargamos el dataset en Polars
print("🔹 Cargando datos en Polars...")
df_polars = pl.read_csv(csv_url)

# Cargamos el mismo dataset en Pandas para comparación
print("🔹 Cargando datos en Pandas...")
df_pandas = pd.read_csv(csv_url)

# Comparación de dimensiones
print(f"\n📊 Dimensiones en Polars: {df_polars.shape}")
print(f"📊 Dimensiones en Pandas: {df_pandas.shape}")

# Exploración básica
print(f"\n🔹 Primeras filas en Polars:")
display(df_polars.head(3))

print(f"\n🔹 Estructura del DataFrame:")
display(df_polars.describe())
```

# 🎯 2. Selección y Filtrado de Datos

Una de las grandes ventajas de Polars es su **sintaxis expresiva y encadenable**, que hace el código más legible comparado con Pandas.

```python
# Comparación de sintaxis: Filtrar clientes de Chile y seleccionar columnas específicas

print("📌 Objetivo: Filtrar clientes de Chile y seleccionar nombre, apellido y compañía\n")

print("🔸 SINTAXIS EN PANDAS (corchetes anidados - menos legible):")
res_pandas = df_pandas[df_pandas["Country"] == "Chile"][["First Name", "Last Name", "Company"]]
display(res_pandas.head(3))

print("\n🔸 SINTAXIS EN POLARS (métodos encadenados - más legible):")
res_polars = (
    df_polars
    .filter(pl.col("Country") == "Chile")  # Filtro de filas
    .select(["First Name", "Last Name", "Company"])  # Selección de columnas
)
display(res_polars.head(3))

print("\n✅ Ambos producen el mismo resultado, pero Polars es más expresivo y fácil de leer")
```

# ⚡ 3. Comparativa de Rendimiento

Uno de los principales puntos fuertes de Polars es su **velocidad excepcional**. Vamos a medir tiempos de ejecución en operaciones de agrupación y agregación.

```python
# Operación: Contar clientes por país (agrupación y agregación)

print("🔸 Midiendo rendimiento en PANDAS...")
start_time = time.time()
agg_pandas = df_pandas.groupby("Country")["Index"].count()
pandas_time = time.time() - start_time
print(f"   Tiempo: {pandas_time:.5f} segundos")

print("\n🔸 Midiendo rendimiento en POLARS...")
start_time = time.time()
agg_polars = (
    df_polars
    .group_by("Country")
    .agg(pl.len().alias("Total_Customers"))  # pl.len() cuenta las filas
)
polars_time = time.time() - start_time
print(f"   Tiempo: {polars_time:.5f} segundos")

# Comparación
print(f"\n{'='*60}")
print(f"⚡ RESULTADO: Polars es {pandas_time / polars_time:.2f}x más rápido")
print(f"{'='*60}")

print(f"\n🔹 Resultado en Polars:")
display(agg_polars.head(10))
```

# 🚀 4. Prueba de Estrés con Grandes Volúmenes

¿Qué pasa cuando realmente manejamos **millones de filas**? Vamos a multiplicar nuestro dataset x50 para simular un escenario de Big Data.

```python
print("🔄 Multiplicando los datos 50 veces...")
print("   Dataset original: 100,000 filas")
print("   Dataset ampliado: 5,000,000 filas\n")

# Multiplicamos los DataFrames
df_pandas_large = pd.concat([df_pandas] * 50, ignore_index=True)
df_polars_large = pl.concat([df_polars] * 50)

print(f"✅ Nuevas dimensiones: {df_polars_large.shape[0]:,} filas x {df_polars_large.shape[1]} columnas\n")

# Medimos Pandas
print("🔸 Midiendo PANDAS con 5M filas...")
start_time = time.time()
agg_pandas_large = df_pandas_large.groupby("Country")["Index"].count()
pandas_time = time.time() - start_time
print(f"   Tiempo: {pandas_time:.5f} segundos")

# Medimos Polars
print("\n🔸 Midiendo POLARS con 5M filas...")
start_time = time.time()
agg_polars_large = (
    df_polars_large
    .group_by("Country")
    .agg(pl.len().alias("Total_Customers"))
)
polars_time = time.time() - start_time
print(f"   Tiempo: {polars_time:.5f} segundos")

# Comparación final
print(f"\n{'='*60}")
if polars_time < pandas_time:
    print(f"🚀 ¡POLARS DOMINA! Es {pandas_time / polars_time:.2f}x más rápido con 5M filas")
else:
    print(f"🐼 Pandas gana esta vez (inusual, pero puede ocurrir)")
print(f"{'='*60}")
```

# 🧠 5. Evaluación Perezosa (Lazy Evaluation)

Una característica **avanzada y poderosa** de Polars es su capacidad de **optimizar automáticamente** las consultas antes de ejecutarlas. Esto se logra mediante `LazyFrame`.

### Concepto de Lazy Evaluation

- **`pl.scan_csv()`** → Crea un `LazyFrame` (no carga datos inmediatamente)
- Se construye un **plan de ejecución** con todas las operaciones encadenadas
- **`collect()`** → Ejecuta el plan de forma optimizada
- Polars analiza y optimiza el plan antes de ejecutarlo

```python
# Creamos una consulta perezosa con pl.scan_csv()
print("🔹 Construyendo consulta perezosa (LazyFrame)...")
query_perezosa = (
    pl.scan_csv(csv_url)  # Lectura perezosa (no carga los datos aún)
    .filter(pl.col("Country") == "Japan")  # Filtrar clientes de Japón
    .group_by("City")  # Agrupar por ciudad
    .agg(pl.len().alias("Clientes_en_Japon"))  # Contar clientes
    .sort("Clientes_en_Japon", descending=True)  # Ordenar descendente
)

print(f"✅ Tipo de objeto: {type(query_perezosa)}")
print("   ⚠️ Los datos AÚN NO se han procesado. Solo tenemos el plan de ejecución.\n")

# Ahora ejecutamos el plan optimizado
print("🔹 Ejecutando con collect()...")
resultado_final = query_perezosa.collect()

print("\n📊 Resultado Final - Top 5 ciudades de Japón con más clientes:")
display(resultado_final.head(5))

print(f"\n✅ Polars optimizó automáticamente la consulta antes de ejecutarla")
print(f"   - Solo leyó las columnas necesarias (Country, City)")
print(f"   - Aplicó filtros antes de cargar todos los datos")
print(f"   - Paralelizó operaciones cuando fue posible")
```
